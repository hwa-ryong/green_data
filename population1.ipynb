```python
import csv
import matplotlib.pyplot as plt

with open('datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    next(data)
    result = [] # 연령별 인구 저장할 리스트

    for row in data:
        if '신림동' in row[0]:
            for i in row[3:]:  # 3번 인덱스 -103(104-1)번 인덱스까지 반복
                result.append(int(i.replace(",", '')))  # 천단위 구분기호 제거
                
print(len(row[3:]))
#print(result)
plt.rc('font', family='Malgun Gothic')
plt.style.use("ggplot")  # 격자 무늬 스타일
plt.title('연령별 인구 통계')
plt.plot(result, 'r')   # y축값
plt.xlabel('연령')
plt.ylabel('인구')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


with open('./datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    next(data)
    result = [] # 연령별 인구 저장할 리스트
    # 지역의 이름 입력받기
    name = input("지역의 이름(읍면동 단위)을 입력해주세요 : ")
    
    for row in data:
        if '평창동' in row[0]:
            for i in row[3:]: # 3번 인덱스 ~ 100번 인덱스
                result.append(int(i.replace(",",'')))

print(len(row[3:]))
# print(result)
plt.rc('font', family='Malgun Gothic')
plt.style.use("ggplot")
plt.title(name + "지역의 인구 구조")
#plt.plot(result, 'r') # y축값
plt.bar(range(101), result) # x축(1-100)
plt.xlabel('연령')
plt.ylabel('인구')
plt.show()
```
